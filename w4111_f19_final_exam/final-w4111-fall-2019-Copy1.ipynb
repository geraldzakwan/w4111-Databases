````python
import pymysql
import logging
import json

default_db_host="localhost"
default_db_user="dbuser"
default_db_password="dbuserdbuser"

def get_new_connection(host, user, password):
    result_conn = pymysql.connect(
        host=host,
        user=user,
        password=password,
        cursorclass=pymysql.cursors.DictCursor,
        autocommit=False)
    return result_conn


midterm_conn = get_new_connection(default_db_host, default_db_user, default_db_password)
````

````python
def tx_commit(conn):
    conn.commit()
    
def tx_rollback(conn):
    conn.rollback()
    
def conn_close(conn):
    conn.close()

def run_q(sql, args=None, fetch=True, cur=None, conn=midterm_conn, commit=True, debug=False):
    '''
    Helper function to run an SQL statement.

    This is a modification that better supports HW1. An RDBDataTable MUST have a connection specified by
    the connection information. This means that this implementation of run_q MUST NOT try to obtain
    a defailt connection.

    :param sql: SQL template with placeholders for parameters. Cannot be NULL.
    :param args: Values to pass with statement. May be null.
    :param fetch: Execute a fetch and return data if TRUE.
    :param conn: The database connection to use. 
    :param cur: The cursor to use. 
    :param commit: Commit after completing transaction.
    :param debug: If true, print some debugging messages.

    :return: A pair of the form (execute response, fetched data). There will only be fetched data if
        the fetch parameter is True. 'execute response' is the return from the connection.execute, which
        is typically the number of rows effected.
    '''

    cursor_created = False
    connection_created = False

    try:

        if conn is None:
            connection_created = True
            conn = get_new_connection(default_db_host, default_db_user, default_db_password)
            cursor_created = True
            cur = conn.cursor()

        if cur is None:
            cursor_created = True
            cur = conn.cursor()

        if args is not None:
            log_message = cur.mogrify(sql, args)
        else:
            log_message = sql

        if debug:
            print("Executing SQL = " + log_message)

        res = cur.execute(sql, args)

        if fetch:
            data = cur.fetchall()
        else:
            data = None

        # Do not ask.
        if commit == True:
            conn.commit()
            
        if cursor_created:
            cur.close()
        if connection_created:
            conn.close()

    except Exception as e:
        if commit == True:
            conn.rollback()
        if cursor_created:
            cur.close()
        if connection_created:
            conn.close()
        raise(e)

    return (res, data)
````

Provide a two or three sentence definition of each of the following concepts, and provide an example.

1. Structured data
<br><br>

<br><br>
3. Unstructured data
<br><br>

<br><br>

Answer:

<a id='2'></a>

1. Structured data <br><br>
Structured data is a data that strictly follows a predefined data model and has a high degree of organization. The model is tabular, containing columns (i.e. the data fields such as person name) and rows (i.e. the records or data sample). Since it has predefined fields, we can easily analyze the data, e.g. we can query by fields (or by a unique identifier called primary key), aggregate some fields or even combines fields from different table/relationship.
<br><br>
Example: Relational data stored in RDBMS (MySQL, PostgreSQL, etc.) and other tabular data such as Excel files.
<br><br>
2. Semi-structured data <br><br>
Semi-structured data is a data that is not defined by the formal model associated with relational database (uses a fix set of columns, uses fix data type for each column, etc.). However, it contains tags to differentiate between elements with different meaning. Using tags, it can also build a hierarchies of records (object inside an object) or fields (e.g. a composite field) within the data.
<br><br>
Example: XML, JSON and YAML documents.
<br><br>
3. Unstructured data <br><br>
Unstructured data is a data that either doesn't have a predefined model (set of columns for example) or is not organized in a predefined manner (stored sequentially for example). Thus, it is hard to search through the data and to extract a specific information from the data as compared to the structured data.
<br><br>
Example: PDF documents, text messages (SMS, WhatsApp, etc.), social media photos, YouTube videos and document-based database such as MongoDB.
<br><br>
4. Metadata <br><br>
Metadata is a data about data. It describes another data (its type, structure, location, etc.) or provides some additional information regarding the data. Almost all kind of data is accompanied by a metadata, mostly for unstructured data (e.g. YouTube videos have a metadata describing the uploader, date uploaded, video category, etc.). Even structured data has metadata such as in the example below.
<br><br>
Example: MySQL stores a metadata that tells us about all the tables in a database schema and the property of each table (its columns, primary key, etc.). We can access this metadata via some commands like "SHOW TABLES" to list all the tables and "DESC <table_name>" to list columns in a table.
<br><br>
5. BLOB <br><br>
BLOB (Binary Large Object) is a data type in DBMS to store a large chunk of data in its binary form. It is usually used to store unstructured data that are not within the standard data types for DBMS (not integer, float, string, etc.) and takes much spaces. BLOB fields allocate space (up to several Giga Bytes) only whenever the content in the field is utilized.
<br><br>
Example: Let say we want to store social media photos and their caption in a database. Since images can't fit into any standard data type in DBMS and they use much spaces, we can use BLOB to store them. We can then store the caption simply as a string. So, we would have a table that looks like : (ID, BLOB_IMAGE, CAPTION).
<br><br>

<span style="font-size:18px;"><u>Q4: Normalization Concepts (5 points)</u></span>

Briefly explain the following concepts. Three or four sentences is sufficient for each concept.
1. Functional dependency.
<br><br>
2. Data redundancy and update anomaly.
<br><br>
3. Decomposition
<br><br>
5. BCNF
<br><br>
5. First Normal Form and atomic domain


Answer:

1. Functional dependency. <br><br>
Functional dependency is a dependency between two attributes. It is usually between the primary key and other non-key attributes. In a table, an attribute P is said to be functionally dependent on Q, if for every Q, the value of Q uniquely determines the value of P. For example, in a table of Columbia student, the field "UNI" should uniquely determine "last_name", "first_name" and "major". In other words, those three columns are functionally dependent to "UNI".
<br><br>
2. Data redundancy and update anomaly. <br><br>
Data redundancy is a condition in a database system where the same piece of data is replicated, i.e. it is stored in two or more different places/records. When data is redundant, update anomaly can occur. The anomaly occurs when we incorrectly changes some existing information. Consider a table -> (bank_customer, branch, branch_address). If the branch_address for a certain branch changes, we need to update all the customer records which are associated the branch. Otherwise, the update anomaly occurs and there will be a data inconsistency. 
<br><br>
3. Decomposition <br><br>
Decomposition is the process of breaking down a relation into smaller relations. The process should be lossless, i.e. it needs to make sure that the information in the original relation can be reconstructed based on the decomposed relations. Decomposition is often used to get rid of data redundancy. Consider the previous example. We can decompose the table by creating a new table that stores (branch, branch_address) and then create a foreign key from the original relation to the new table.
<br><br>
5. BCNF <br><br>
When a table has more than one candidate key, anomalies may result even though the relation is in 3NF. Boyce-Codd normal form is a special case of 3NF. A relation is in BCNF if, and only if, every determinant is a candidate key.
<br><br>
5. First Normal Form and atomic domain <br><br>
Atomic domain means that every value of in a field is an indivisible unit. A database is said to be in the first normal form (1NF) if it satisfies these two conditions: 1. It contains only atomic values 2. There are no repeating groups.

1. Give three advantages of a B+ Tree index relative to a hash index. <br>
    - B+ Tree index is optimized for range query retrieval using inequality operator (e.g. searching for records ranging from key value A to key value B). Hash index on the other hand will suffer because we can't sort based on the hashed value. Hash index only supports equality operator.
    - B+ Tree index can speed up ORDER BY operations, meanwhile hash index is not able to do that with the same reason as above.
    - With a B+ Tree index, we can search by any leftmost prefix of the key, useful when dealing with string based key. Meanwhile, hash index can only search based on the whole key.
<br><br>
2. What is a covering index? Why use it? <br><br>
A covering index is a non-clustered index which includes all columns that are referred in a query (all columns in the SELECT, JOIN, WHERE, GROUP BY, etc.). Covering index is useful when we frequently query based on some non-key columns that are not indexed. When using covering index, we avoid a further lookup to retrieve any non-key data. Normally, without covering index, we still have to scan sequentially to match the non-key columns after using the non-clustered index. With covering index, we avoid that further lookup as all columns are in the non-clustered index.
<br><br>
3. What issue does a write intensive workload cause for B+ Tree indexes? <br><br>
Too many insertions and deletions will cause a considerable amount of overhead for B+ Tree Indexes. This is because it maintains pointers not only from parent to child but also sequential pointers in the last level (leaf nodes). Most of the time, an insertion or deletion would cause a rearrangement of the nodes and sometimes extra space. This becomes worse when we insert a record with the same key (if alllowed) or when we delete a node such that some nodes must be heavily rearranged, i.e. some nonleaf nodes are moved bottom up.
<br><br>
4. VERY briefly describe an LSM Tree. <br><br>
LSM Tree ("Log Structured Merge Tree") is a way to store data with robustness against high insert and update rates. It consists of two or more levels, just like any other tree. The simplest versions consist of two levels: C0 and C1. C0 is small and stored entirely in memory, whereas C1 is huge and stored on disk. Records are inserted (or updated) in C0 (often called memtable, implemented using, for example, an AVL Tree). If the insertion is performed when C0 is full, a contiguous segment of entries is removed from C0 and merged into C1 on disk. C1 comprises of many immutable log segments accompanied by index, e.g. hash index, for quick lookup. C1 is generally implemented using SSTables (string sorted table).
<br><br>
5. VERY briefly explain index key compression, how it works, its benefits and when it is applicable. <br><br>
Index key compression reduces the size of index file by compressing repeating key values. It splits the index key into two parts: 
    - The prefix column (which are potentially shared across multiple key values); 
    - The suffix column (which is unique to every index key). The same prefix columns will be stored once. <br><br>
Some benefits beside storage efficiency: 
    - Better cache efficiency when loading index to memory
    - Fewer leaf blocks and shallow tree (if using tree based index)
    - Fewer logical IOs as size is smaller <br><br>
Index key compression is applicable (or makes sense) when:
    - The index has repeated leading values
    - The index is a composite index of 2 or more columns
    - Compression defined for a maximum of n-1 columns  (where n is the number of columns in the index).  That is, the last column cannot be compressed.

Answer:

1. Why is querying  ```lahman2019clean.annual_performance_summary_copy``` so much faster than querying ```lahman2019clean.annual_summary?``` <br><br>
```annual_performance_summary_copy``` is a copy table meanwhile ```lahman2019clean.annual_summary``` is a view. When querying from a copy table, we operate directly on the records that are already stored within the copy table. Meanwhile, when querying on a view, we instead need to execute all the operations needed to create the view from the actual table, starting from the ```batting_summary``` to the ```annual_summary```. View is a virtual table afterall, thus there is no record stored in it. Thus, querying the ```annual_performance_summary_copy``` is much faster than querying ```annual_summary``` as we don't have to requery the records.
<br><br>
2. If views are so much slower than copy tables, give the primary reason why a database design might choose a view? <br><br>
Copy tables are not associated with the base table, i.e. changes in the base table won't be reflected on the copy table (unless we define a way to do so). Thus, when we query on the copy table, we might get an outdated data. Meanwhile, when we use a view, we will always query based on the current condition of the base table. Hence, if the base table is write intensive (thus it is an overhead to reflect changes from the base table to the copy table) and the view operations are light (doesn't involve multiple joins), then view is more desirable over copy table.
<br><br>
3. What are three general reasons/general benefits of using views compared to just executing the SQL on underlying tables. <br><br>
    - View provides an abstraction to simplify the query. View can draw data from several tables and present a single aggregated table. Thus, it simplifies the information, hides the complexity and turns multi-table queries into single-table queries.
    - View enforces logical data independence and consistency. A view can present a consistent structure of the database, even if the underlying tables are split, restructured, or renamed. Thus, changes in underlying table don't affect its user (or an application program that uses it).
    - View can act as a security mechanism, i.e. restrict what columns and/or records can be accessed by a certain group of users. This allows surfacing only the data that a user needs to see.

4. Explain, in two or three sentences, how you handle reflecting changes in the base tables in the data in the copy table. <br><br>

    - For UPDATE operation only, check if the updated columns are used in any of the view (batting_summary, pitching_summary, fielding_summary, appearances_summary, annual_summary). If there isn't any view using them, no further action is needed. Otherwise (or if the operation is INSERT or DELETE), proceed to the next steps below.
    - Drop the copy table, i.e. execute ```DROP TABLE lahman2019clean.annual_performance_summary_copy```
    - Finally, recreate the copy table using ```CREATE TABLE lahman2019clean.annual_performance_summary_copy AS SELECT * from annual_summary;```
